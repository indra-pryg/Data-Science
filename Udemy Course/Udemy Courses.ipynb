## 8. Pelatihan apa saja yang di publikasikan pada tahun 2015

```python
df.head(2)
```

```python
df.dtypes
```

```python
df.published_timestamp = pd.to_datetime(df.published_timestamp)
```

```python
df.drop(columns='Year', axis=1, inplace=True)
```

```python
df['Year'] = df.published_timestamp.dt.year

df.head(2)
```

```python
df[df.Year == 2015].Year.count()
```

## 9. Jumlah Pelatihan setiap tahunnya

```python
df.groupby('Year').count()
```

```python
yy = df.groupby('Year').course_title.describe()
```

```python
yy.iloc[:, :1].plot.bar()
plt.show();

yy.iloc[:, :1]
```
